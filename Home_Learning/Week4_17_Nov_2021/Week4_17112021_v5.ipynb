```python
import altair as alt


source = data_df1

brush = alt.selection(type='interval')

points = alt.Chart(source,title="Work in progress vs Incentive").mark_point(filled = True).encode(
    x='incentive:Q',
    y=alt.Y('wip:Q', title='Work in progress (wip)'),
    #y='wip:Q',
    color=alt.condition(brush, 'day:N', alt.value('lightgray'))
).add_selection(
    brush
)

bars = alt.Chart(source).mark_bar().encode(
    y='day:N',
    color='day:N',
    x=alt.X('count(day):Q', title='Amount of days')
    #x='count(day):Q'
    
).transform_filter(
    brush
)

points & bars
```
